## Multiplot

Now we want to introduce useful ways to create multiplots. It would be necessary to compare two or more plots. 

There are two different functions that we are going to discuss them in the following blocks:

. plt.subplots()

. plt.subplot()

Let's start with plt.subplots(). This function helps you to create multiplots.

See : https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/subplots_demo.html

The return of plt.subplots() function is a set of axes. Here an axis means the position of a plot in the figure.

So consider there is a grid consists of different plots. The return axes of plt.subplots() specify a grid that looks like a matrix and has several single plots inside itself. 

For more information see the example in : https://www.tutorialspoint.com/matplotlib/matplotlib_subplots_function.htm

Let's start...

plt.subplots() without arguments returns a figure and a single axis. It is just like the simple plotting that we dicussed before. Use axis.plot(array1, array2) to show the plot in the specific axis (positin of grid). Because there is not any argument in plt.subplots(), there is just one axis right now. Pay attention! In this way there is no need to write plt.show().

```python
x = np.arange(0, 10 , 0.1)
y = x**2
# write your code here
axs = plt.subplot()
axs.plot(y)
# write your code here
```

On the other hand, when you're going to create multiplots with plt.subplots(), you should pay attention to the arguments. The first two optional arguments of plt.subplots() define the number of rows and columns of the subplot grid. When you just use one argument, they will stack in one column and the returned axis is a 1D numpy array containing the list of created axis.

In this part we want to use just one argument to plt.subplots(). As I told you before, when you just use one argument, they will stacking in one column. But the axis that returned from this function, is an 1d array. Each of the elements of axes set are specific axis for specific position of that column. Stack the plot of the sine and cosine function for array t in one column. Again plt.subplots() returns a figure within the axis set.

```python
from numpy import pi
t = np.arange(0, 4*pi , 0.05)
# write your code here

fig, axs = plt.subplots()
axs.plot(np.sin(t))
fig, axs = plt.subplots()
axs.plot(np.cos(t))



# write your code here
```

In the block code above, we can use " fig, (ax1, ax2) = plt.subplots(2) " instead of " fig, axis = plt.subplots(2) "
... So each axis is individually stored in a variable and we don't need to use 1d array axis.

And also if you need to have plots in two directions as a matrix (not just in one direction), pay attention to the return axis array that has the same shape as your arguments were.

Plot y1 = (u^2)+1 in row 1 and column 1, y2 = (u^2)-1 in row 1 and column 2, y3 = (u+1)^2 in row 2 and column 1, y4 = (u-1)^2 in row 2 and column 2.

```python
u = np.arange(-10,10,0.2)
# write your code here
fig, axis = plt.subplots(2,2)
y1 = (u**2)+1
y2 = (u**2)-1
y3 = (u+1)**2
y4 = (u-1)**2
axis[0, 0].plot(y1)
axis[0, 1].plot(y2)
axis[1, 0].plot(y3)
axis[1, 1].plot(y4)
# write your code here

axis[0,0].set_ylim(-5,10)
axis[0,0].grid(True, which='both')
axis[0,0].axhline(y=0, color='k')
axis[0,0].axvline(x=0, color='k')
axis[0,1].set_ylim(-5,10)
axis[0,1].grid(True, which='both')
axis[0,1].axhline(y=0, color='k')
axis[0,1].axvline(x=0, color='k')
axis[1,0].set_ylim(-5,10)
axis[1,0].grid(True, which='both')
axis[1,0].axhline(y=0, color='k')
axis[1,0].axvline(x=0, color='k')
axis[1,1].set_ylim(-5,10)
axis[1,1].grid(True, which='both')
axis[1,1].axhline(y=0, color='k')
axis[1,1].axvline(x=0, color='k')
```

Let's do some exercises on the plt.subplot() function.

plt.subplot() and plt.subplots() are different in usage but are almost the same in their name. What is their difference? 

In plt.subplots() function that we discussed before, you can use this single function to create a figure with several subplots with only one line of code. As you learned, we have an 1d or 2d array of axes as return object and it consists of the position of plots. But plt.subplot() creates only a single subplot axis at a specified grid position. This means it will require several lines of code to achieve the same result as matplot.pyplot.subplots() did in a single line.

See : https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplot.html

Now use plt.subplot() to plot

y1 = e^(x) 

and    

y2 = log(x)

in one row. First argument of plt.subplot() indicates number of "rows" in whole figure, second one indicates number of "columns" in whole figure and third argument indicates the index of this plot as using indexes in a matrix! Pay attention to use plt.show().

```python
x = np.arange(0.01, 5, 0.1)
# write your code here


y1 = np.e**(x)

y2 = np.log(x)

axs = plt.subplot()
axs.plot(y1)
plt.show()

axs = plt.subplot()
axs.plot(y2)

plt.show()

# write your code here
```

## Bar Plot

Bar plot is one of the most useful plotting forms that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. 

I have defined a dictionary about the number of different T-shirt colors in a clothing store; and I've put the keys and values in two separate lists. Use plt.bar(List1 , List2) to see the bar plot of these dataset. Set the color argument of plt.bar() to 'aqua' and width argument of rectangulars to 0.5.

```python
d = {'red':20, 'green':15, 'blue':30, 'white':35}
keys = list(d.keys()) 
values = list(d.values()) 
fig = plt.figure(figsize = (10, 5)) 
# write your code here
plt.bar(keys,values)
plt.show()

# write your code here
```
