### np.arange()  --- np.linspace()

Let's work with two useful functions. 1- np.arange()  - 2- np.linspace()

For the first one, np.arange() creates sequence of numbers. This is the exact form of this function : np.arange(start, stop, step).... Start is start of interval. Stop is end of interval and Step is spacing between values. It returns a ndarray.


Pay attention! You have to mention "stop" argument whereas you can use default values for "start" and "step". The default value of start is 0 and default value of step is 1.

Create an array whose elements start in 0, end in 4 and they increase one by one. So you will have {0, 1, 2, 3, 4} ... Use np.arange() and just use "stop" argument!!! Be careful... set "stop" argument to real stop number plus 1 ! 

```python
# write your code here
print(np.arange(4+1))
# write your code here
```

```python
# write your code here
print(np.arange(10.0,11.0,0.20))
# write your code here
```

Another useful function to create arrays in numpy is np.linspace() . The exact form of this function is : np.linspace(start , stop , num). It returns an array with sequence of numbers. "Start" is the starting value of the sequence. "Stop" is the end value of the sequence and "num" is number of samples to generate.


Pay attention! You have to mention "start" and "stop" argument whereas ou can use default value for "num" that is 50. There is another important argument called "endpoint" that if it is True, "stop" is the last sample. Otherwise, it is not included. Default is True.

Let's see an example. Generate an array whose elements are between 0 and 10. we want to have 5 numbers in this array.

```python
# write your code here

print(np.linspace(0,10,5))
# write your code here
```

It's easy to do basic operations such as adding or subtracting using numpy. If you have n dimensional matrix, it doesn't matter.

As you can see, dimensions of these operand matrices must be the same. But there is an important exception that calls Broadcasting.

Broadcasting describes how numpy treats arrays with different shapes during arithmetic operations. Actually the smaller array is “broadcast” across the larger array so that they have compatible shapes. The simplest broadcasting example occurs when an array and a scalar value are combined in an operation. You can see other examples here: https://numpy.org/doc/stable/user/basics.broadcasting.html 

In this example, array J and scalar value "b" are defined. Just do multiplication on them.

```python
a1 = np.ones((3,1))
b1 = np.random.rand(3,1)
J = a1 + b1
print(J)
b = 6
# write your code here
print(J*b)
# write your code here
```

## Index and iteration

### 1d array

We just want to use indexes to make changes on 1d arrays

Create an array with 9 elements between 10 and 50. Return first 3 elements with indexing and then set the value of odd indexes to 0.

```python
# write your code here

S= np.linspace(10,50,9)[0:3]
for i in np.arange(len(S)):
    if i%2 != 0:
        S[1]=0

# write your code here
print(S)
```

Use indexes to reverse array S

```python
# write your code here
print(S[:None:-1])
# write your code here
```
