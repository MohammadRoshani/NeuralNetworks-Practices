Suppose we want to extend our function in a way that gets at most 3 numbers, calculates root square of each of them and returns their results.

Note 2 : The main purpose of this part is learning how to return multi-numbers .


```python
#write your code here 

def square_root_calculator_base(num):
    if num <0:
        return -1
    else: 
        return  np.sqrt(num)
    
def square_root_calculator(num1 = 0, num2 = 0, num3 = 0):
    return  square_root_calculator_base(num1),square_root_calculator_base(num2),square_root_calculator_base(num3)
         

#write your code here 

result = square_root_calculator(-16)
print(result)

result = square_root_calculator(-16 , 25)
print(result)

result = square_root_calculator(-16 , 25 , 9)
print(result)
```

##  3. Dictionaries:

A dictionary is like an address-book where you can find the address or contact details of a person by knowing only his/her name i.e. we associate keys (name) with values (details). Note that the key must be unique just like you cannot find out the correct information if you have two persons with the exact same name.

Note that you can use only immutable objects (like strings) for the keys of a dictionary but you can use either immutable or mutable objects for the values of the dictionary. This basically translates to say that you should use only simple objects for keys

### 3.1 

Creating dictionaries : dictionaries can be created with { key1 : value1  , key2 : value 2 , ...} 

Size of dictionaries : dictionaries size can be obtained with len() e.g. len(dictionary)

Access elements : dictionaries elements can simply be obtained from their keys e.g. : dictionary [ key1 ] = value1

Now we want to construct a dictionary that maps every country to its capital. e.g. : capital[ "Iran" ]  = " Tehran " .


```python
countries = ["Iran" , "France" , "England" , "China" , "Russia "]
capitals  = ["Tehran" , "Paris" , "London" , "Beijing" , "Moscow "]
capital_dict = {}
#write your code here

for i in range(len(countries)):
    capital_dict[countries[i]]=capitals[i]

#write your code here
print(len(capital_dict))
print(capital_dict)
```

### 3.2 

Delete from a dictionary : We can delete an item with the specified key name. e.g. : del dictionary[key]

Now please delete "Iran" from the previous dictionary

```python
#write your code here
del capital_dict["Iran"]
#write your code here
print(capital_dict)
```

### 3.3
